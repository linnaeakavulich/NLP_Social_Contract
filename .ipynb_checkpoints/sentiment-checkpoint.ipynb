````python
hobbes_sents['sent_str'] = hobbes.groupby(SENTS).term_str.apply(lambda x: x.str.cat(sep=' '))
hobbes_sents['html_str'] = hobbes.groupby(SENTS).html.apply(lambda x: x.str.cat(sep=' '))
````

````python
def sample_sentences(df):
    rows = []
    for idx in df.sample(10).index:

        valence = round(df.loc[idx, emo], 4)     
        t = 0
        if valence > t: color = '#ccffcc'
        elif valence < t: color = '#ffcccc'
        else: color = '#f2f2f2'
        z=0
        rows.append("""<tr style="background-color:{0};padding:.5rem 1rem;font-size:110%;">
        <td>{1}</td><td>{3}</td><td width="400" style="text-align:left;">{2}</td>
        </tr>""".format(color, valence, df.loc[idx, 'html_str'], idx))

    display(HTML('<style>#sample1 td{font-size:120%;vertical-align:top;} .sent-1{color:red;font-weight:bold;} .sent1{color:green;font-weight:bold;}</style>'))
    display(HTML('<table id="sample1"><tr><th>Sentiment</th><th>ID</th><th width="600">Sentence</th></tr>'+''.join(rows)+'</table>'))
````

````python
sample_sentences(hobbes_sents)
````

## Sentiment Analysis Using Vader

````python
analyser = SentimentIntensityAnalyzer()
````

````python
hobbes_vader_cols = hobbes_sents.sent_str.apply(analyser.polarity_scores).apply(lambda x: pd.Series(x))
hobbes_vader = pd.concat([hobbes_sents, hobbes_vader_cols], axis=1)
````

````python
w = int(hobbes_vader.shape[0] / 5)
hobbes_vader[['pos','neg']].rolling(w).mean().plot(figsize=(25,5))
hobbes_vader[['neu']].rolling(w).mean().plot(figsize=(25,5))
hobbes_vader[['compound']].rolling(w).mean().plot(figsize=(25,5))
````

## Polarity by document

````python
# For TOKENS
OHCO = ['book_id', 'chap_num', 'para_num', 'sent_num', 'token_num']
BOOKS = OHCO[:1]
CHAPS = OHCO[:2]
PARAS = OHCO[:3]
SENTS = OHCO[:4]
````

````python
CHAPS
````

````python
VOCAB = pd.read_csv('VOCAB.csv')
TOKENS = pd.read_csv('CORPUS_tfidf.csv').set_index(OHCO).sort_index()
LIB = pd.read_csv('LIB.csv').drop(columns="Unnamed: 0").set_index('book_id')
````

````python
def create_bow(CORPUS, bag, item_type='term_str'):
    BOW = CORPUS.groupby(bag+[item_type])[item_type].count().to_frame('n')
    return BOW
````

````python
TOKENS
````

````python
BOW = TOKENS.reset_index().groupby(CHAPS+['term_str', 'tfidf'])['term_str'].count().to_frame('n')
````

````python
BOW = BOW.reset_index().set_index(CHAPS+['term_str'])
````

````python
SALEX = pd.read_csv(salex_csv).set_index('term_str')
SALEX.columns = [col.replace('nrc_','') for col in SALEX.columns]
````

````python
SALEX['sentiment'] = SALEX.positive - SALEX.negative
````

````python
SALEX.head(3)
````

````python
# lexicon columns
emo_cols = "anger anticipation disgust fear joy sadness surprise trust sentiment".split()
````

````python
# SALEX + VOCAB table
V = pd.concat([VOCAB.reset_index().set_index('term_str'), SALEX], join='inner', axis=1)
````

````python
V.head(3)
````

````python
# VOCAB + BOW:
B = BOW.join(V[['max_pos'] + emo_cols], on='term_str', rsuffix='_v').dropna()
````

````python
B
````

````python
# weight emotions by TFIDF
for col in emo_cols:
    B[col] = B[col] * B.tfidf
````

````python
B.head(3)
````

````python
EMO_BOOKS = B.groupby(['book_id'])[emo_cols].mean()
EMO_CHAPS = B.groupby(['book_id','chap_num'])[emo_cols].mean()
````

````python
EMO_BOOKS.index = LIB.title 
````

````python
EMO_BOOKS.plot.barh(figsize=(15,30));
````

````python
locke = EMO_BOOKS.loc[['The Second Treatise of Government', 'An Essay Concerning Humane Understanding',
                      'An Essay Concerning Humane Understanding, Volume 2']].copy()
````
